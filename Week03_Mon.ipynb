```python
import bisect

def subset_sum_mitm(arr, target):
    n = len(arr)
    mid = n // 2

    left = arr[:mid]
    right = arr[mid:]

    sum_left = []
    sum_right = []

    # Generate all subset sums of left half
    for mask in range(1 << len(left)):
        s = 0
        for i in range(len(left)):
            if mask & (1 << i):
                s += left[i]
        sum_left.append(s)

    # Generate all subset sums of right half
    for mask in range(1 << len(right)):
        s = 0
        for i in range(len(right)):
            if mask & (1 << i):
                s += right[i]
        sum_right.append(s)

    # Sort right half sums for binary search
    sum_right.sort()

    # Check for target sum
    for s in sum_left:
        need = target - s
        idx = bisect.bisect_left(sum_right, need)
        if idx < len(sum_right) and sum_right[idx] == need:
            return True

    return False


# Main Driver Code
t = int(input())

for _ in range(t):
    n, S = map(int, input().split())
    arr = list(map(int, input().split()))

    if subset_sum_mitm(arr, S):
        print("YES")
    else:
        print("NO")
```
