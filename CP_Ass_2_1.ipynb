```python
def maxSubArray(a, l, r):
    if l == r:
        return a[l]
    m = (l + r) // 2
    left = maxSubArray(a, l, m)
    right = maxSubArray(a, m + 1, r)
    s = 0
    lm = rm = -10**18
    for i in range(m, l - 1, -1):
        s += a[i]
        lm = max(lm, s)
    s = 0
    for i in range(m + 1, r + 1):
        s += a[i]
        rm = max(rm, s)
    return max(left, right, lm + rm)
T = int(input())
for _ in range(T):
    N = int(input())
    A = list(map(int, input().split()))
    print(maxSubArray(A, 0, N - 1))
```
