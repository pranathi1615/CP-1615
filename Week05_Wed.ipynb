```python
# Segment Tree for Range Min and Max Queries

import math

# Input number of sensors
n = int(input("Enter number of sensors: "))
arr = list(map(int, input(f"Enter {n} temperature readings: ").split()))

# Initialize segment trees
size = 2 * (2 ** math.ceil(math.log2(n))) - 1
seg_min = [float('inf')] * size
seg_max = [float('-inf')] * size

# Build Segment Tree for Min
def build_min(node, start, end):
    if start == end:
        seg_min[node] = arr[start]
    else:
        mid = (start + end) // 2
        build_min(2*node+1, start, mid)
        build_min(2*node+2, mid+1, end)
        seg_min[node] = min(seg_min[2*node+1], seg_min[2*node+2])

# Build Segment Tree for Max
def build_max(node, start, end):
    if start == end:
        seg_max[node] = arr[start]
    else:
        mid = (start + end) // 2
        build_max(2*node+1, start, mid)
        build_max(2*node+2, mid+1, end)
        seg_max[node] = max(seg_max[2*node+1], seg_max[2*node+2])

# Range Min Query
def range_min(node, start, end, l, r):
    if r < start or end < l:
        return float('inf')
    if l <= start and end <= r:
        return seg_min[node]
    mid = (start + end) // 2
    left = range_min(2*node+1, start, mid, l, r)
    right = range_min(2*node+2, mid+1, end, l, r)
    return min(left, right)

# Range Max Query
def range_max(node, start, end, l, r):
    if r < start or end < l:
        return float('-inf')
    if l <= start and end <= r:
        return seg_max[node]
    mid = (start + end) // 2
    left = range_max(2*node+1, start, mid, l, r)
    right = range_max(2*node+2, mid+1, end, l, r)
    return max(left, right)

# Update operation
def update(node, start, end, idx, value):
    if start == end:
        arr[idx] = value
        seg_min[node] = value
        seg_max[node] = value
    else:
        mid = (start + end) // 2
        if idx <= mid:
            update(2*node+1, start, mid, idx, value)
        else:
            update(2*node+2, mid+1, end, idx, value)
        seg_min[node] = min(seg_min[2*node+1], seg_min[2*node+2])
        seg_max[node] = max(seg_max[2*node+1], seg_max[2*node+2])

# Build both trees
build_min(0, 0, n-1)
build_max(0, 0, n-1)

# Example queries
print(range_min(0, 0, n-1, 2, 6))  # RangeMin(2,6)
print(range_max(0, 0, n-1, 1, 5))  # RangeMax(1,5)
update(0, 0, n-1, 3, 27)           # Update(3,27)
print(range_min(0, 0, n-1, 2, 6))  # RangeMin(2,6) after update
```
