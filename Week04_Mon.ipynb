```python
class FenwickTree:
    def __init__(self, n):
        self.n = n
        self.tree = [0] * (n + 1)

    def update(self, index, delta):
        while index <= self.n:
            self.tree[index] += delta
            index += index & -index

    def query(self, index):
        total = 0
        while index > 0:
            total += self.tree[index]
            index -= index & -index
        return total


# Number of days
n = int(input("Number of days: "))

# Attendance points
attendance = list(map(int, input("Daily attendance points: ").split()))

fenwick = FenwickTree(n)

# Build Fenwick Tree
for i in range(n):
    fenwick.update(i + 1, attendance[i])

# Query 1: total attendance till Day 4
print("Total attendance till Day 4 =", fenwick.query(4))

# Update: Day 3 attendance from 1 â†’ 2
day = 3
new_value = 2
delta = new_value - attendance[day - 1]
attendance[day - 1] = new_value
fenwick.update(day, delta)

# Query 2: total attendance till Day 4
print("After update, total attendance till Day 4 =", fenwick.query(4))
```

```python
class FenwickTree:
    def __init__(self, n):
        self.n = n
        self.tree = [0] * (n + 1)

    def update(self, index, delta):
        while index <= self.n:
            self.tree[index] += delta
            index += index & -index

    def query(self, index):
        total = 0
        while index > 0:
            total += self.tree[index]
            index -= index & -index
        return total

n = int(input())
sales = list(map(int, input().split()))

fenwick = FenwickTree(n)

# Build tree
for i in range(n):
    fenwick.update(i + 1, sales[i])

q = int(input())

for _ in range(q):
    op = input().split()

    if op[0] == "QUERY":
        d = int(op[1])
        print(f"Total sales revenue till Day {d} = {fenwick.query(d)}")

    elif op[0] == "UPDATE":
        i = int(op[1])
        new_val = int(op[2])

        delta = new_val - sales[i - 1]
        sales[i - 1] = new_val
        fenwick.update(i, delta)
```
