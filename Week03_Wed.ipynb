```python
data = """1
5
1 3 5 7 9
4
2 1 3
1 2 10
2 1 3
2 0 4
"""

data = data.strip().split('\n')
idx = 0

def input():
    global idx
    val = data[idx]
    idx += 1
    return val


T = int(input())

for _ in range(T):
    N = int(input())
    arr = list(map(int, input().split()))

    size = 1
    while size < N:
        size <<= 1

    seg = [0] * (2 * size)

    # Build
    for i in range(N):
        seg[size + i] = arr[i]

    for i in range(size - 1, 0, -1):
        seg[i] = seg[2 * i] + seg[2 * i + 1]

    Q = int(input())
    for _ in range(Q):
        q = list(map(int, input().split()))

        if q[0] == 1:
            _, i, x = q
            i += size
            seg[i] = x
            while i > 1:
                i //= 2
                seg[i] = seg[2 * i] + seg[2 * i + 1]
        else:
            _, L, R = q
            L += size
            R += size
            ans = 0
            while L <= R:
                if L & 1:
                    ans += seg[L]
                    L += 1
                if not (R & 1):
                    ans += seg[R]
                    R -= 1
                L //= 2
                R //= 2
            print(ans)
```
