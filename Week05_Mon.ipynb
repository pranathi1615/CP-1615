```python
n = int(input("Enter number of servers: "))

# Initialize parent array (each server is its own parent initially)
parent = [i for i in range(n)]

# Find with path compression
def find(u):
    if parent[u] != u:
        parent[u] = find(parent[u])
    return parent[u]

# Union operation
def union(u, v):
    root_u = find(u)
    root_v = find(v)
    if root_u != root_v:
        parent[root_v] = root_u

# Check connectivity
def connected(u, v):
    return find(u) == find(v)

# Input connections
m = int(input("Enter number of connections: "))
print("Enter the connections (u v) one per line:")
for _ in range(m):
    u, v = map(int, input().split())
    union(u, v)

# Input queries
q = int(input("Enter number of connectivity queries: "))
print("Enter the queries (u v) one per line:")
for _ in range(q):
    u, v = map(int, input().split())
    print("YES" if connected(u, v) else "NO")
```
